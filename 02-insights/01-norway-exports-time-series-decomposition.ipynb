```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.tsa.stattools import adfuller, kpss

import warnings
warnings.filterwarnings('ignore')
```

(0) Load

```python
df_norway_exports = pd.read_parquet('df_nsc_2025_until_w18.parquet')
df_norway_exports = df_norway_exports.loc[df_norway_exports['Country'].isin(['TOTALT'])]
df_norway_exports['Week'] = df_norway_exports['Week'].astype('int32')
df_norway_exports['Year-Week'] = df_norway_exports['Year'].astype('string') + df_norway_exports['Week'].astype('string')
print(df_norway_exports.info())
df_norway_exports.head()
```

(1) Time Series Overall Exports Volume

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_norway_exports, x='Week', y='Volume (KG)', hue='Year')
plt.title('Overall Norway Exports Volume 2023 - 2025')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_norway_exports, x='Year-Week', y='Volume (KG)')
plt.title('Overall Norway Exports Volume 2023 - 2025')
plt.show()
```

Segn la tendencia de las exportaciones, es ms parecido a un modelo aditivo que multiplicativo. Debido a que la tendencia no aumento desde un ao a otro. 
Para efectos del anlisis, utilizaremos los aos 2023 y 2024, debido a que hasta la fecha, el 2025 ha tenido un comportamiento un poco anmalo, debido a las buenas condiciones biolgicas, es decir, mejor cosecha.

(2) Time Series Decomposition

- Additive Model

```python
df_norway_exports_2023_2024 = df_norway_exports.loc[df_norway_exports['Year'].isin([2023, 2024])]
df_norway_exports_2023_2024 = df_norway_exports_2023_2024.groupby(
                                                                    by=['Year', 'Week'], as_index=False
).agg(
        VolumeKG = ('Volume (KG)', 'sum')
).sort_values(by=['Year', 'Week'], ascending=True)

df_norway_exports_2023_2024['year_week_date'] = pd.to_datetime(
                                                                df_norway_exports_2023_2024['Year'].astype(str) 
                                                                + 
                                                                df_norway_exports_2023_2024['Week'].astype(str).str.zfill(2) 
                                                                + '1', 
                                                                format='%G%V%u')

df_volumes = df_norway_exports_2023_2024.set_index('year_week_date')
df_volumes = df_volumes['VolumeKG']
```

```python
# yearly
additive_decomposition = seasonal_decompose(df_volumes, model='additive', period=52)
additive_decomposition.plot()
plt.show()
```

```python
# semester
additive_decomposition = seasonal_decompose(df_volumes, model='additive', period=26)
additive_decomposition.plot()
plt.show()
```

```python
resultado = adfuller(np.log(df_volumes)) # se convirti en log para quitar el efecto no estacionario

print(f'Estadstico ADF: {resultado[0]}')
print(f'Valor p: {resultado[1]}')
print('Valores crticos:')
for clave, valor in resultado[4].items():
    print(f'  {clave}: {valor}')
    
# Interpretacin bsica
if resultado[1] < 0.05:
    print("La serie es estacionaria (rechazamos H0)")
else:
    print("La serie no es estacionaria (no rechazamos H0)")
```

```python
#serie_log = np.log(df_volumes)
serie_log = df_volumes

resultado_kpss = kpss(serie_log, regression='ct')

print(f'Estadstico KPSS: {resultado_kpss[0]}')
print(f'Valor p: {resultado_kpss[1]}')
print('Valores crticos:')
for clave, valor in resultado_kpss[3].items():
    print(f'  {clave}: {valor}')

if resultado_kpss[1] < 0.05:
    print("La serie **no es estacionaria** (rechazamos H0)")
else:
    print("La serie **es estacionaria** (no rechazamos H0)")
```

### Conclusin: 
Las exportaciones de noruega a nivel anual, poseen una tendencia en forma de U invertida, debido a que existe una baja durante la primera parte del ao, para luego ser aumenta en segunda parte, por mejores condiciones del mercado, el cual hace que la planning del cultivo sea para cosechar en esa parte. Por lo tanto tiene sentido la forma de U. 
Y a nivel semestral, que es el otro horizonte que decid descomponer, consta de dos Ues. una normal y la segunda invertida. Reflejando que al principio del ao la cosecha tiene a la baja, debido a que no es un perodo de fuerte demana, para luego aumentar (u invertida) en donde las condiciones de mercado llegan a su peak. Los dos ciclos tienen un inicio y fin, por eso se reflejan en forma de U. 
Los residuos, a nivel anual, son prcticamente nulos, debido a que no existe una mayor diferencia entre un ao y otro, sera mejor si tuviese a lo menos 5 aos de exportacin. Y, a nivel semestral, se ven un poco distorsionados, en donde debe ser por fechas puntuales del comercio, como vacaciones, festividades o algn evento biolgico, etc. 
Y desde un punto de vista estadstico, tiene sentido que no sea estacionaria (test ADF), debido a que existen fluctuaciones por eventos del negocio (vacaciones, festividades, etc), y desde el test KS, dice que s es estacionaria pero convirtindola a logaritmo, y por muy poco. Y al quitarle la tendencia, refleja lo mismo.


Ahora, al no estar conforme con los resultados, considerando que sta informacin es lo que pasa en la realidad, decid verlo en un horizonte por trimestre, con el objetivo de ver qu qu identifica el modelo, ms que el contexto de negocio. 

```python
# trimester: classical decomposition
additive_decomposition = seasonal_decompose(df_volumes, model='additive', period=12)
additive_decomposition.plot()
plt.show()
```

```python
# trimester: STL decommposition
stl = STL(df_volumes, period=12)
result = stl.fit()

trend = result.trend
seasonal = result.seasonal
resid = result.resid

result.plot()
plt.show()
```

Considerando que es un modelo aditivo, se aplic la descomposicin en TSA, considerando una estacionalidad por trimestre. Finalmente el grfico qued ms sensata que con el modelo tradicional, ya que refleja patrones estaciones de subidas y bajas no tan pronunciadas y constantes.
